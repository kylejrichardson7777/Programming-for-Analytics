1. Create a list with 3 student dictionaries 

```python
students= [{'name': 'Alice', 'grade': 80},
           {'name': 'Bob','grade': 70}, 
           {'name': 'Charlie', 'grade': 60} ] 
```

```python

for s in students:
    print(s['name'])
```

2. Print the average grade

```python
avg_g = 0
for g in students:
    avg_g += g['grade']
avg_grade = avg_g / len(students)
print(avg_grade)
```

3. print the names of the students who scored above average

```python
for g in students:
    if g["grade"]> avg_grade:
        print(g["name"])
        
```
