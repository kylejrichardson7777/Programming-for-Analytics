## Task 1: Even or Odd Number Checker

```python
number = input("even or odd checker :")
if int(number) % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
```

## Task 3: Password Strength Checker

```python
password = input("password checker:")

if len(password) >= 8:
    if any(c.isdigit() for c in password):
        if any(not c.isalnum() for c in password):
            print("strong password")
        else:
            print("password does not contain special character")
    else:
        print("password does not contain digits")
else:
    print("password too short")
```

## Task 4: FizzBuzz

```python
for n in range(51):
    if n % 3 == 0 and n % 5 == 0:
        print("fizzbuzz")
    elif n % 3 == 0:
        print("fizz")
    elif n % 5 == 0:
        print("buzz")
    else:
        print(n)
```

## Task 5: Temperature Conversion Table

```python
def C_to_F(start,end):
    
    for c in range(start, end+1):
        f = c*9/5 +32
        print(f"{c}C = {f}F")
        
C_to_F(0,10)
```

## Task 6: Factorial with Input Validation

```python
number = input("Insert your number: ")

if not number.isdigit():
    print("Error: please enter a positive integer.")
    
else:
    number = int(number)
    factorial = 1
    for i in range(1, number + 1):
        factorial *= i
    print(f"The factorial of {number} is {factorial}.")

```

## Task 7: Prime Number Checker

```python
def prime_checker(number):
    number = int(number)
    
    if number < 2:
        print(f"{number} is not a prime number")
        return
    
    for i in range(2, number):
        if number % i == 0:
            print(f"{number} is not a prime number")
            return
            
    print(f"{number} is a prime number")
```
