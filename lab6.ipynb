```python
import pandas as pd, csv
```

```python
dic = {}
with open("malazan_bridgeburners.csv", "r") as mala:
    dictionary =csv.DictReader(mala)
    data = [d for d in dictionary]
data[5]
df = pd.DataFrame(data)
df
```

```python
df["Battles_Fought"] = df["Battles_Fought"].astype(float)
df["Battles_Fought"].mean()
```

```python
df["Avg_Damage"] = df["Avg_Damage"].astype(int)
df["Avg_Damage"].median()
```

```python
df["Healing_Spells"] = df["Healing_Spells"].astype(int)
df["Healing_Spells"].max()
```

```python
df["Injuries_Sustained"] = df["Injuries_Sustained"].astype(int)
df[(df["Battles_Fought"] >7) & (df["Avg_Damage"] > 85)]
```

```python
new_df = df[(df["Battles_Fought"] >7) & (df["Avg_Damage"] > 85)].copy()
new_df.sort_values(by = "Avg_Damage", ascending = False)
```

```python
new_df["Battle_Effciency"] = new_df["Avg_Damage"] * new_df["Battles_Fought"] / new_df["Injuries_Sustained"]
new_df
```

```python
new_df.sort_values(by = "Battle_Effciency", ascending = False)
```

```python
be_summary =new_df.groupby("Role").mean("Battle_Effciency")
be_summary
```

```python
be_summary.to_csv("bridgeburners_summary.csv", index = False)
```
